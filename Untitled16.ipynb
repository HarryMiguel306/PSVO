```python
#importation des bibliothèques nécessaires pour réaliser la segmentation avec kmeans
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import shutil
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
```

```python
#Création d'un dossier contenant 100 images

source_folder = "/input/ISIC2018_Task1-2_Training_Input"
destination_folder = "/destination input"

os.makedirs(destination_folder, exist_ok=True)

image_files = sorted([f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))])

# Copie des 100 premières images dans le dossier de destination
for i, file_name in enumerate(image_files[:100]):
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(destination_folder, file_name)
    shutil.copy(source_path, destination_path)
    print(f"Copié : {file_name} ({i + 1}/100)")

print("100 premières images copiées avec succès.")
```

```python
#Création d'un dossier contenant 100 images

source_folder = "/input_groundtruth/ISIC2018_Task1_Training_GroundTruth"
destination_folder = "/destination groundthruth"

os.makedirs(destination_folder, exist_ok=True)

image_files = sorted([f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))])

# Copie des 100 premières images dans le dossier de destination
for i, file_name in enumerate(image_files[:100]):
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(destination_folder, file_name)
    shutil.copy(source_path, destination_path)
    print(f"Copié : {file_name} ({i + 1}/100)")

print("100 premières images copiées avec succès.")
```

```python
#prétraitement
def preprocess(image):
    # Convertir l'image en niveaux de gris
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Redimensionnement
    resized = cv2.resize(gray, (128, 128))
    # Normalisation
    normalized = resized / 255.0
    return normalized


def preprocess_dataset(image_dir, img_size=(128, 128)):
    images = []

    # Trier les fichiers pour correspondre les images et les masques
    image_files = sorted(os.listdir(image_dir))

    for img_file in image_files:
        # Charger et prétraiter l'image
        img_path = os.path.join(image_dir, img_file)
        image = cv2.imread(img_path)
        if image is not None:
            preprocessed_image = preprocess(image)
            if preprocessed_image is not None:
                images.append(preprocessed_image)

    # Convertir les listes en tableaux NumPy
    images = np.array(images)

    return images


images = preprocess_dataset('/destination input')
print(images.shape)
```

```python
#kmeans
def kmeans_segmentation(image, k=2):
    # Redimensionner l'image
    image_reshaped = image.reshape(-1, 1)

    # Appliquer K-means
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(image_reshaped)
    labels = kmeans.labels_

    # Reconstruire l'image segmentée
    segmented_image = labels.reshape(image.shape)

    return segmented_image


def segment_dataset(image_dir, mask_dir, k=2):
    segmented_images = []

    # Trier les fichiers pour correspondre les images et les masques
    image_files = sorted(os.listdir(image_dir))
    mask_files = sorted(os.listdir(mask_dir))
    for img_file, mask_file in zip(image_files, mask_files):
        # Charger et prétraiter l'image
        img_path = os.path.join(image_dir, img_file)
        mask_path = os.path.join(mask_dir, mask_file)
        image = cv2.imread(img_path)
        mask = cv2.imread(mask_path)
        if image is not None and mask is not None:
            preprocessed_image = preprocess(image)
            if preprocessed_image is not None:
                segmented_image = kmeans_segmentation(preprocessed_image, k)
                segmented_images.append((segmented_image, mask))

    return segmented_images


segmented_images = segment_dataset('/destination input', '/destination groundthruth')
```
